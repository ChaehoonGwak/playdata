# 9월 3일 ~ 9일 금융 경제 기사 수집하여 트렌드 분석하기 

```python
import requests
from bs4 import BeautifulSoup as BS
import os
from datetime import datetime, date, timedelta
import time
```

```python
# 검색조건 : 전날로부터 일주일치 검색
now = datetime.now()
searchDate = [int(now.strftime("%Y%m%d"))-i for i in range(1,2)]
```

```python
searchDate
```

```python
head = {
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36 Edg/93.0.961.38"
}
```

```python
rootFolder = "news"
if not os.path.isdir("./{}".format(rootFolder)): os.mkdir("./{}".format(rootFolder))

pageNum = 2 # default 100page
exhead = ""
for searchDay in searchDate:
    if not os.path.isdir("./{}/{}".format(rootFolder, searchDay)): os.mkdir("./{}/{}".format(rootFolder, searchDay))
    for i in range(1,pageNum):
        head = {
        'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36 Edg/93.0.961.38"
        }
        r = requests.get("https://news.naver.com/main/list.naver?mode=LS2D&sid2=263&sid1=101&mid=shm&date={}&page={}".format(searchDay, i), headers=head)
        bs =  BS(r.text, "lxml")
        urlList = []
        [urlList.append(x['href']) for x in bs.find("div", class_="list_body newsflash_body").findAll("a") if x['href'] not in urlList]
        for url in urlList:
            head = {
            'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36 Edg/93.0.961.38"
            }
            r2 = requests.get(url, headers=head)
            bs2 = BS(r2.text, "lxml")
            head = bs2.find("h3").text
            body = bs2.find("div", class_="_article_body_contents").text.split("\n")[8]
            #if head == exhead: break
            f = open("./{}/{}/{}.txt".format(rootFolder, searchDay, head), "a")
            f.write(body)
        
            
        #if head == exhead: break    
        exhead = head
        
```
