```python
def fibo(n):
    if n == 0 :
        return 0
    elif n == 1 or n == 2:
        return 1
    else: return fibo(n-1) + fibo(n-2)
    
print(fibo(9))
```

```python
def fibo2(n):
    cur, next_ = 0, 1
    for x in range(n):
        cur, next_ = next_, cur + next_
    return cur
fibo2(40)
```

```python
%%time
fibo(40)
```

```python
def fib3(m, n):
    if m[n] == 0:
        m[n] = fib3(m, n-1) + fib3(m, n-2)
    return m[n]
```

```python
%%time
n = 40
m = [0] * n
m[0], m[1] = 1, 1
print(fib3(m, n-1))
```

```python
mTotal = []
fTotal = []
for roots, dirs, files in os.walk("./이름_생년_성별_10000"):
    if not os.path.isdir("./남성"):
        os.mkdir("./남성")
    if not os.path.isdir("./여성"):
        os.mkdir("./여성")
    for file in files:
        if file.split("_")[2][0] in ['1', '3']:
            with open(roots + "/" + file) as f:
                for x in f:
                    mTotal.append(x.replace(",", ""))
            shutil.move(roots + "/" + file, "./남성" + file)
        else:
            with open(roots + "/" + file) as f:
                for x in f:
                    fTotal.append(x.replace(",", ""))
            shutil.move(roots + "/" + file, "./여성" + file)
```
